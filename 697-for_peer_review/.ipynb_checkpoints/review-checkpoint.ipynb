```python
#just choose the picture you like
import time
import ntpath
import numpy as np
from PIL import Image
from os.path import join, exists
from keras.models import model_from_json
import matplotlib.pyplot as plt
samples_dir = "../outputtest/"
if not exists(samples_dir): os.makedirs(samples_dir)
model_h5 = 'model_15320_.h5'
model_json = 'model_15320_.json'
with open(model_json, "r") as json_file:
    loaded_model_json = json_file.read()
funie_gan_generator = model_from_json(loaded_model_json)
funie_gan_generator.load_weights(model_h5)
img_path='test/1.jpg'
im = Image.open(img_path).resize((256, 256))
inp_img=np.array(im).astype(np.float32)
print("the picture we test is: ", img_path)
im =(inp_img/127.5)-1.0
im = np.expand_dims(im, axis=0) 
gen = funie_gan_generator.predict(im)
gen=np.uint8((gen+1.0)*127.5)
gen_img = gen[0]
img_name = ntpath.basename(img_path)
out_img = gen_img.astype('uint8')
a1=Image.fromarray(out_img).save(join(samples_dir, img_name))
print("\nWe Save the output images in {0}\n".format(samples_dir))
```

```python
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg 
import numpy as np
img1 = mpimg.imread(img_path)
img2 = mpimg.imread(join(samples_dir, img_name))
```

```python
print("the original picture is: ")
plt.imshow(img1)
plt.show()
plt.close()
im2 = Image.open(join(samples_dir, img_name))
print("the output picture is: ")
plt.imshow(img2)
plt.show()
plt.close()
```

```python
from __future__ import division
import numpy as np
import math
from scipy.ndimage import gaussian_filter
from skimage.metrics import structural_similarity
import cv2
ssim = structural_similarity(img1,img2, multichannel=True) 
diff = np.array(img2, dtype=np.float64) - np.array(img1, dtype=np.float64)
diff = diff.flatten('C')
rmse = math.sqrt(np.mean(diff ** 2.) )
if(rmse==0):
    psnr=100
else:
   psnr=20*math.log10(255.0/rmse) 
print("PSNR is: ", psnr)
print("SSIM is: ", ssim)
```
